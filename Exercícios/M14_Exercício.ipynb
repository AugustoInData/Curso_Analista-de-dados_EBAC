## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

import seaborn as sns
import matplotlib.pyplot as plt

# Carregar os dados
diamonds = data #variável utilizada na célula anterior

# Criar o barplot com a média de preços por corte
plt.figure(figsize=(8, 6))
sns.barplot(x="cut", y="price", data=diamonds, ci=None)
plt.title("Preço Médio por Tipo de Corte de Diamante")
plt.xlabel("Tipo de Corte")
plt.ylabel("Preço Médio (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

**Insight do gráfico 1**: ...

Ao analisar o gráfico de barras, percebo que os diamantes com corte "Fair" apresentam, em média, um preço mais alto do que os cortes "Good", "Very Good", "Premium" e até mesmo "Ideal". Isso me chamou a atenção, pois cortes de melhor qualidade costumam ser considerados mais valiosos. A partir disso, entendo que o preço do diamante não depende apenas do tipo de corte. Outros fatores, como o peso em quilates, a cor e a clareza, provavelmente têm um papel importante na formação do preço — e podem estar mais presentes nos diamantes classificados com corte "Fair".

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 2

flights = sns.load_dataset("flights")

import seaborn as sns
import matplotlib.pyplot as plt


# Filtrar apenas os dados do mês de dezembro
dez = flights[flights['month'] == 'Dec']

# Mostrar o gráfico
plt.figure(figsize=(10,6))
sns.lineplot(x='year', y='passengers', data=dez, marker='o')
plt.title('Número de Passageiros em Dezembro por Ano (1949–1960)')
plt.xlabel('Ano')
plt.ylabel('Número de Passageiros')
plt.tight_layout()
plt.show()
```

**Insight do gráfico 2**: ...

Observando o gráfico, percebo que o número de passageiros no mês de dezembro apresenta um crescimento constante ao longo dos anos, refletindo o aumento da demanda no final do ano, possivelmente por causa das férias e das festas de fim de ano. Esse aumento consistente me mostra como dezembro é um mês importante para o setor aéreo, e que as companhias podem aproveitar esse padrão para planejar melhor suas operações e estratégias comerciais nessa época.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3

flights = sns.load_dataset("flights")

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Filtrar para os anos entre 1949 e 1959
data_filtrada = flights[(flights['year'] >= 1949) & (flights['year'] <= 1959)]

# Agrupar por mês e calcular a média de passageiros
media_mensal = data_filtrada.groupby('month')['passengers'].mean().reset_index()

# Para manter a ordem correta dos meses, convertemos 'month' em categorical com ordem
meses_ordenados = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
media_mensal['month'] = pd.Categorical(media_mensal['month'], categories=meses_ordenados, ordered=True)
media_mensal = media_mensal.sort_values('month')

# Mostrar o gráfico
plt.figure(figsize=(10,6))
sns.lineplot(x='month', y='passengers', data=media_mensal, marker='o')
plt.title('Média do Número de Passageiros por Mês (1949–1959)')
plt.xlabel('Mês')
plt.ylabel('Número Médio de Passageiros')
plt.tight_layout()
plt.show()

```

**Insight do gráfico 3**: ...

Ao analisar os dados de 1949 a 1959, percebo que há um padrão muito claro de sazonalidade no número de passageiros ao longo dos meses. Os meses de verão, especialmente julho e agosto, apresentam consistentemente os maiores picos de movimento, enquanto os meses de inverno, como fevereiro e novembro, tendem a ter menor volume. Esse comportamento se repete ano após ano, o que me mostra que a demanda por voos segue um ciclo previsível. Essa informação é valiosa, pois reforça a importância do planejamento de voos, preços e infraestrutura com base na sazonalidade.
